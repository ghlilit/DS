```python
# Creating some new features
df['R_Home_country'] = df['R_HomeTown'].str.split(' ').str[-1]
df['R_Location_country'] = df['R_Location'].str.split(' ').str[-1]

df['B_Home_country'] = df['B_HomeTown'].str.split(' ').str[-1] 
df['B_Location_country'] = df['B_Location'].str.split(' ').str[-1]
```

```python
df['R_Home_country'].replace('', np.nan, inplace=True)
df['B_Location_country'].replace('', np.nan, inplace=True)
df['B_Home_country'].replace('', np.nan, inplace=True)
df['R_Location_country'].replace('', np.nan, inplace=True)
```

```python
columns=['B_HomeTown','B_Location','B_Home_country','B_Location_country',
        'R_HomeTown','R_Location','R_Home_country','R_Location_country']

percent = 0.7
for column in columns:
    l = len(df[column].unique())
    non_common_categories = df[column].value_counts(dropna=False)[int(l*percent):].index
    for c in non_common_categories:
        df[column].replace(c, 'other', inplace=True)
```

# Bin the data for columns R_Weight and B_Weight.

```python
sum(df['R_Weight'].isna())
```

```python
def binning(df, columns, bin_count, labels):
    for i,column in enumerate(columns):
        min_value = df[column].min()
        max_value = df[column].max()
        bins = np.linspace(min_value, max_value, bin_count[i]+1)
        df[column+'_bins'] = pd.cut(df[column], bins=bins, labels=labels[i], include_lowest=True)
```

```python
binning(
    df,
    columns = ['R_Weight','B_Weight'], 
    bin_count = [3,3],
    labels = [['lightweight','midweight','heavyweight'],
              ['lightweight','midweight','heavyweight']])
```

# Handling missing values

```python
df_int = df.select_dtypes(include=["integer"])
```

```python
df_int.isna().sum()
#no missing values in integer columns
```

```python
float_cols = df.select_dtypes(include=["floating"]).columns.values
```

```python
df.select_dtypes(include=["floating"]).isna().sum().sort_values()
```

```python
threshold = int(df.shape[0]*0.4)
df.drop(columns=float_cols[df[float_cols].isna().sum()>threshold], inplace = True)
```

```python
df.select_dtypes(include=["floating"]).isna().sum().sort_values()
```

```python
def fill(df, columns, method='mean'):
    for column in columns:
        if method == 'mode':
            df[column].fillna(df[column].mode()[0], inplace = True)
        else:
            df[column].fillna(df[column].mean(), inplace = True)
```

```python
columns = df.select_dtypes(include=["floating"]).isna().columns.values
fill(df, columns[:170])
fill(df, columns[170:], method = 'mode')
```
